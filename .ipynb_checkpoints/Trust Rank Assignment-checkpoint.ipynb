### Create Nodes and Edges and build the graph.
- Each Node corresponds to a Buyer and Seller.
- The Edges correesponds to transactions between the buyers and sellers.
- We also identify and map the Nodes (Buyes and Sellers) who are colluding.

```python
import pandas as pd
import collections
import threading

from numpy import mat, eye, zeros, ones, linalg, arange
import random
import matplotlib.pyplot as plt
```

```python
def create_graph_nodes():
    # initialize empty lists for buyers and sellers
    buyer = []
    seller = []

    # read in the data from the CSV file
    df = pd.read_csv('Payments.csv')

    # extract unique buyer IDs from the 'Buyer ID' column and append them to the buyer list
    buyers = df["Receiver"].unique()
    for bb in buyers:
        buyer.append(bb)

    # extract unique seller IDs from the 'Seller ID' column and append them to the seller list
    sellers = df["Sender"].unique()
    for ss in sellers:
        seller.append(ss)

    # combine the buyer and seller lists into a single list of unique nodes
    # by converting them to sets, concatenating them, and then converting
    # back to a sorted list
    total_nodes = sorted(set(list(set(seller)) + list(set(buyer))))

    # convert the list of nodes to a list of strings with a newline character before each node
    total_nodes_str = [str(u) if i==0 else "\n" + str(u) for i, u in enumerate(total_nodes)]

    # write the list of node strings to a file named 'nodes_unique.txt'
    with open("nodes_unique.txt", "w") as f:
        f.writelines(total_nodes_str)
```

```python
def create_malicious_nodes():
    # read in the data from the CSV file containing malicious entity IDs
    df = pd.read_csv("bad_sender.csv")

    # extract the IDs of the malicious entities from the 'Bad ID' column and sort them
    bad_entity = df["Bad Sender"].sort_values()

    # convert the list of bad entity IDs to a list of strings with a newline character before each ID
    bad = [str(b) if i ==0 else "\n" + str(b) for i, b in enumerate(bad_entity)]

    # write the list of bad entity ID strings to a file named 'nodes_unique_bad.txt'
    with open("nodes_unique_bad.txt", "w") as f:
        f.writelines(bad)
```

```python
def create_graph_edges():
    # read in the data from the CSV file
    df = pd.read_csv('Payments.csv')

    # group the data by seller ID, buyer ID, and value, and count the number of occurrences
    # of each combination of these three variables
    df = df.groupby(['Sender', 'Receiver', 'Amount']).size().reset_index().rename(columns={0: 'count'})

    # extract the seller IDs, buyer IDs, and weights from the grouped data
    seller = df["Sender"]
    buyer = df["Receiver"]
    weight = df["Amount"]

    # combine the seller IDs, buyer IDs, and weights into a list of strings representing edges
    # in the graph, with each string in the form "buyer ID,seller ID,weight\n"
    write_list = [str(buyer[i]) + "," + str(seller[i]) + "," + str(weight[i]) + "\n" for i in range(len(buyer))]

    # write the list of edge strings to a file named 'edge_unique.txt'
    with open("edge_unique.txt", "w") as f:
        f.writelines(write_list)
```

```python
create_graph_nodes()
create_malicious_nodes()
create_graph_edges()
```

### Implementation of Pregel Algorithm

```python
class Vertex():
    def __init__(self, id, multi, value, out_vertices):
        # Initialize the vertex with a unique ID, a value (initially set to None), 
        self.id = id
        self.value = value
        
        # A multiplier (which is a value that can be used to weight the vertex)
        self.multi = multi
        self.out_vertices = out_vertices

        # initialize the incoming and outgoing message lists to be empty
        self.incoming_messages = []

        # A list of outgoing vertices (i.e., the vertices that this vertex is connected to)
        self.outgoing_messages = []

        # set the vertex to active (i.e., it can participate in computations)
        self.active = True

        # initialize the superstep to 0 (this will be incremented by the Pregel algorithm 
        # during each iteration of computation)
        self.superstep = 0
```

```python
class Worker(threading.Thread):
    def __init__(self, vertices):
        # Initialize the worker thread with a list of vertices to process.
        threading.Thread.__init__(self)
        self.vertices = vertices

    def run(self):
        # This method will be called when the worker thread starts running.
        self.superstep()

    def superstep(self):
        # Loop through all the vertices that this worker thread is responsible for.
        for vertex in self.vertices:
            # If the vertex is active (i.e., it can participate in computation)
            if vertex.active:
                # Call the update() method on the vertex, which will process incoming messages
                # and update the vertex's state
                vertex.update()
```

```python
class Pregel():

    def __init__(self, vertices, num_workers):
        self.vertices = vertices
        self.num_workers = num_workers

    def run(self):
        # Runs the Pregel instance.
        self.partition = self.partition_vertices()
        while self.check_active():
            self.superstep()
            self.redistribute_messages()

    def partition_vertices(self):
        # Returns a dict with keys 0,...,self.num_workers-1 representing the worker threads.
        # The corresponding values are lists of vertices assigned to that worker
        partition = collections.defaultdict(list)
        for vertex in self.vertices:
            partition[self.worker(vertex)].append(vertex)
        return partition

    def worker(self, vertex):
        # Returns the id of the worker that vertex is assigned to.
        return hash(vertex) % self.num_workers

    def superstep(self):
        """
        Completes a single superstep.  

        Note that in this implementation, worker threads are spawned,
        and then destroyed during each superstep.  This creation and
        destruction causes some overhead, and it would be better to
        make the workers persistent, and to use a locking mechanism to
        synchronize.  The Pregel paper suggests that this is how
        Google's Pregel implementation works.
        """
        workers = []
        for vertex_list in self.partition.values():
            worker = Worker(vertex_list)
            workers.append(worker)
            #print(f" Worker started with vertices list : {[v.id for v in vertex_list]}")
            worker.start()
        for worker in workers:
            worker.join()

    def redistribute_messages(self):
        # Updates the message lists for all vertices
        for vertex in self.vertices:
            vertex.superstep +=1
            vertex.incoming_messages = []

        for vertex in self.vertices:
            for (receiving_vertix,message) in vertex.outgoing_messages:
                receiving_vertix.incoming_messages.append((vertex,message))

    def check_active(self):
        #Returns True if there are any active vertices, and False otherwise."""
        return any([vertex.active for vertex in self.vertices])
```

### Implementation of TrustRank using pregel classes implemented above

```python
num_workers = 4

# Vertex class for Trustrank algorithm
class TRVertex(Vertex):
    def __init__(self, id, value, multi, out_vertices, damping_factor=0.85, iterations=50):
        # Initialize the TRVertex object with a unique ID, initial value, out-vertices, and other parameters
        Vertex.__init__(self, id, value, multi, out_vertices)
        self.num_supersteps = iterations
        self.damping_factor = damping_factor

    def update(self):
        # Check if the vertex has exceeded the maximum number of supersteps
        if self.superstep < self.num_supersteps:
            # Compute the sum of incoming messages
            messages_sum = 0
            for (vertex, message) in self.incoming_messages:
                messages_sum += message
            
            # Update the vertex's value using the trust rank formula
            if len(self.incoming_messages) > 0:
                self.value = (1 - self.damping_factor) * self.multi + self.damping_factor * messages_sum

            # Compute the outgoing messages for each out-vertex
            outgoing_message = self.value
            sum_vertex = 0
            for vertex in self.out_vertices:
                sum_vertex += vertex[1]
            self.outgoing_messages = [(vertex[0], (outgoing_message * vertex[1]) / sum_vertex) for vertex in self.out_vertices]
        else:
            # If the vertex has exceeded the maximum number of supersteps, mark it as inactive
            self.active = False
```

```python
# Return a list(graph nodes) after file read
def read_nodes(filename):
    with open(filename,'r') as f:
        return [int(n) for n in f.readlines()]

# Intialiaze value for each node
# For Trustrank it is same of all nodes : 1 / No of nodes
def startTrustRank(nodes, malicious_node_filename):

    # Create the set of malicious buyer/seller nodes
    malicious_node = set(read_nodes(malicious_node_filename))

    # Normalize the value of the initial score assigned to the vector.
    rank = [1.0/len(malicious_node) if n in malicious_node else 0 for n in nodes]
    return rank

# Update out going vertices for each vertex using edges.
def updateOutBoundVertices(vertices, node_dict, edges_file, bad_nodes_file):

    malicious_nodes = set(read_nodes(bad_nodes_file))
    with open(edges_file,'r') as f:
        for edge in f.readlines():

            edge_list = edge.strip('\n').split(",")
            frm, to, wt = int(edge_list[0]), int(edge_list[1]), float(edge_list[2])
            
            if frm == to:
                continue

            v = node_dict[frm]; flag = 0

            for i,(x,y) in enumerate(v.out_vertices):
                if x.id == to:
                    v.out_vertices[i] = (x, y+wt)
                    flag=1
                    break
            if not flag:
                v.out_vertices.append((node_dict[to], wt))
                    

    for v in vertices:
        if len(v.out_vertices) == 0:
            for b in malicious_nodes:
                # Assign equal weight of 1 to each node 
                v.out_vertices.append((node_dict[b],1))
```

```python
def displayResults(matrix):
    # convert matrix to list and flatten
    mat_to_list = matrix.tolist()
    flat_list = []
    for i in mat_to_list:
        for k in i:
            flat_list.append(k)
    flat_list.sort()
    
    # create list of bins for histogram
    blist = []
    w = 0.006
    while w < 0.042:
        blist.append(w)
        w += 0.006
    
    # create figure and subplots
    f = plt.figure(figsize=(15,4))
    ax1 = f.add_subplot(121)
    ax2 = f.add_subplot(122)

    # plot histogram of trust rank scores
    ax1.hist(flat_list, bins=blist, color="red", edgecolor='black', linewidth=1.2)
    ax1.set_xlabel("Trust Rank")
    ax1.set_ylabel("Count")
    ax1.set_title("Trust Rank vs Count")
    ax1.grid()

    # plot trust rank scores for each node
    ax2.plot(list(range(len(flat_list))), flat_list)
    ax2.set_xlabel("Node Index")
    ax2.set_ylabel("Trust Rank Score")
    ax2.set_title("Node Index v.s. Trust Score")
    ax2.grid()
```

```python
# Compute Trustrank using pregel by assing set of vertices to thread.
def pregelTrustRankScore(vertices):
    pObj = Pregel(vertices,num_workers)
    pObj.run()
    nodeids = [vertex.id for vertex in pObj.vertices]
    trustrank = [vertex.value for vertex in pObj.vertices]
    matrix = mat([vertex.value for vertex in pObj.vertices]).transpose()
    return (nodeids, trustrank, matrix)

nodes_file = "nodes_unique.txt"
bad_nodes_file = "nodes_unique_bad.txt"

# Nodes of graph
nodes = read_nodes(nodes_file); num_vertices = len(nodes)

# Intial ranks of nodes in the order of nodes
nodes_intial_ranks = startTrustRank(nodes,bad_nodes_file)

#Intialize the vertices with modified Vertex class
new_vertices = [] ; node_dict = {}
for i in range(len(nodes)):
    new_vertices.append(TRVertex(nodes[i], nodes_intial_ranks[i], nodes_intial_ranks[i], [], 0.85, 50))
    node_dict[nodes[i]]= new_vertices[i]

#Updating vertex out vertices to send the messages
edge_file = "edge_unique.txt"
updateOutBoundVertices(new_vertices, node_dict, edge_file, bad_nodes_file)

#Trustrank computation
nodeids, trustrank, matrix = pregelTrustRankScore(new_vertices)

ndf = pd.DataFrame({
    'NodeIDs': nodeids,
    'TrustRank': trustrank
})

print(ndf.head(10))

displayResults(matrix)
```
